# Data Preprocessing
- Remove missing
- Data Normalization (min max scalling)

```python
# removing missing values
df.dropna(axis=0,inplace=True)
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:,1:].values
y = df.iloc[:,0].values
```

```python
X.shape
```

# Min Max Scalling 


```python
X.min() , X.max()
```

```python
Xnorm = X / X.max()
```

```python
Xnorm
```

```python
Xnorm.shape
```

```python
# female = 1, male = 0
y_norm = np.where(y=='female',1,0)
```

```python
# save x and y in numpy zip
np.savez('./data/data_10000_norm.npz',Xnorm,y_norm)
```
